# Task 1

```python
test_01 = """RL

AAA = (BBB, CCC)
BBB = (DDD, EEE)
CCC = (ZZZ, GGG)
DDD = (DDD, DDD)
EEE = (EEE, EEE)
GGG = (GGG, GGG)
ZZZ = (ZZZ, ZZZ)"""

test_02 = """LLR

AAA = (BBB, BBB)
BBB = (AAA, ZZZ)
ZZZ = (ZZZ, ZZZ)"""
```

```python
def read_input(input):
  split_lines = input.split('\n')
  # first line are the directions
  directions = split_lines[0]
  # then we have a blank line
  # and then we have the triplets
  map = {}
  for i in range(2, len(split_lines)):
    res = re.findall(r'\w+', split_lines[i])
    map.update({res[0]: (res[1], res[2])})
  return directions, map
```

```python
print(read_input(test_01))
```

```python
def map_L_and_R(letter):
  if letter == 'L':
    return 0
  return 1

def travel_the_map(directions, map):
  location = 'AAA'
  dir_iter = 0
  num_step = 0
  while location != 'ZZZ':
    loc_tuple = map[location]
    location = loc_tuple[map_L_and_R(directions[dir_iter])]
    dir_iter += 1
    num_step += 1
    if dir_iter == len(directions):
      dir_iter = 0
  return num_step

def get_answer(input):
  directions, map = read_input(input)
  steps = travel_the_map(directions, map)
  return steps
```

```python
print(get_answer(test_01))
print(get_answer(test_02))
```

```python
def find_nodes_ending_with_letter(map, letter):
  keys = map.keys()
  result = [x for x in keys if x.endswith(letter)]
  return result

def check_if_all_end_with_letter(list_locs, letter = 'Z'):
  result = [x for x in list_locs if x.endswith(letter)]
  return result == list_locs
```

```python
# this is taking a lot of time.... probably better not to brute force it
def ghost_travel_the_map(directions, map):
  locations = find_nodes_ending_with_letter(map, 'A')
  print('Number of starting locations:', len(locations))
  dir_iter = 0
  num_step = 0
  while not check_if_all_end_with_letter(locations, 'Z'):
    if num_step % 1000000 == 0:  # million step
      print(num_step)
    dir = map_L_and_R(directions[dir_iter])  # same for all locs
    # now map all locations at the same time
    for i in range(len(locations)):
      loc_tuple = map[locations[i]]
      locations[i] = loc_tuple[dir]
    dir_iter += 1
    num_step += 1
    if dir_iter == len(directions):
      dir_iter = 0
  return num_step

def get_answer_ghost(input):
  directions, map = read_input(input)
  steps = ghost_travel_the_map(directions, map)
  return steps
```

```python
print(get_answer_ghost(test_03))
```
