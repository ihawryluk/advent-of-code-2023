```python
def make_list_of_locs(iterator):
  result = []
  for Match in iterator:
    result.append(Match.start())  # those are all single chars
  return result

def make_list_locs_numbers(iterator):
  result = []
  for Match in iterator:
    result.append(Match.span())  # those are mostly multi-digit
  return result
```

```python
def parse_lines_not_edges(line_str1, line_str2, line_str3):
  result = 0
  # do something else if this is the first or last line
  # or like make fake lines with just dots
  # otherwise
  # for the middle line find the numbers
  numbers_locs = re.finditer(r'(\d+)', line_str2)
  numbers_locs_list = make_list_locs_numbers(numbers_locs)
  # for the same line find the special chars
  # line_same_locs = re.finditer(r'([^a-zA-Z0-9.])', line_str2)
  # line_same_locs_list = make_list_of_locs(line_same_locs)
  # # for the first line find the special chars
  # line_before_locs = re.finditer(r'([^a-zA-Z0-9.])', line_str1)
  # line_before_locs_list = make_list_of_locs(line_before_locs)
  # # for the third line find the special chars
  # line_after_locs = re.finditer(r'([^a-zA-Z0-9.])', line_str3)
  # line_after_locs_list = make_list_of_locs(line_after_locs)

  for loc in numbers_locs_list:
    # print(loc)
    # extract the string around the numbers:
    n = len(line_str1)
    new_string = line_str1[max(loc[0]-1,0):min(loc[1]+1,n)]  # is this gonna break at a single digit????
    new_string += line_str2[max(loc[0]-1,0):min(loc[1]+1,n)]
    new_string += line_str3[max(loc[0]-1,0):min(loc[1]+1,n)]
    # print(new_string)
    # check if there are any special chars (not diits and not .) in the new_string
    if re.search(r'([^a-zA-Z0-9.])', new_string):  # a special symbol was found
      result += int(line_str2[loc[0]:loc[1]])

  return result
```

```python
def parse_first_line(line_str2, line_str3):
  return parse_lines_not_edges(line_str3, line_str2, line_str3)

def parse_last_line(line_str2, line_str3):
  return parse_lines_not_edges(line_str2, line_str3, line_str2)
```

```python
def parse_whole_input(input):
  result = 0
  # split into lines
  split_lines = input.split('\n')
  # print(split_lines[0])
  result += parse_first_line(split_lines[0], split_lines[1])
  # print(result)
  result += parse_last_line(split_lines[-2], split_lines[-1])
  # print(result)
  for i in range(1,len(split_lines)-1):
    # print(i)
    # print(split_lines[i])
    result += parse_lines_not_edges(split_lines[i-1], split_lines[i], split_lines[i+1])
  return result
```
