```python
def calibration_line_with_words(line_txt):
  # try to find first digit
  dig_1_loc = re.search(r'\d+', line_txt)[1]
  dig_1 = line_txt[dig_1_loc]

  # try to find first word-digit before that
  # no digit found
  if dig_1_loc == 0:
    dig_1_loc == -1

  for i in range(len(word_digits)):
    index = line_txt[0:dig_1_loc].find(word_digits[i])
    if index != -1:   # a word-digit found
      dig_1 = str(i + 1)
      break


  # try to find the last digit after that
  dig_2_loc = re.search(r'\d+', line_txt[::-1])[1]
  dig_2 = line_txt[::-1][dig_2_loc]  # CHECK THIS

  # find the word-digits from the end
  for i in range(len(word_digits)):
    index = line_txt[0:-(dig_2_loc)].rfind(word_digits[i])
    if index != -1:   # a word-digit found
      dig_2 = str(i + 1)
      break

  return dig_1[0] + dig_2[0]   # these [0]s to get the first digits only
```
